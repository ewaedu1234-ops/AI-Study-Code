```python
import requests

def get_weather(city_name):
    API_KEY = "4ae57fd056a83f59e7561791b26b4d3a"
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={API_KEY}"
    response = requests.get(url)
    data = response.json()
    return data['weather'][0]['main']
```

```python
# 이제 이름만 부르면 끝납니다!
print("서울:", get_weather("Seoul"))
print("부산:", get_weather("Busan"))
print("제주:", get_weather("Jeju"))
```
