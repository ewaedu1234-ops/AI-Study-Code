```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 시연용 CSV 데이터 읽기 (앞서 만든 weather_data.csv 사용)
# 만약 파일이 없다면 위에서 생성한 파일을 먼저 실행해주세요.
df = pd.read_csv("weather_data.csv")

# 2. 그래프 한글 깨짐 방지 설정 (Windows 기준)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 3. 그래프 그리기
plt.figure(figsize=(10, 5)) # 그래프 크기 설정

# 기온 데이터 선 그래프 그리기
plt.plot(df['날짜'], df['기온'], marker='o', color='red', linestyle='-', linewidth=2, label='기온 (°C)')

# 그래프 꾸미기
plt.title('지난 7일간의 기온 변화 추이', fontsize=16)
plt.xlabel('날짜', fontsize=12)
plt.ylabel('기온 (°C)', fontsize=12)
plt.xticks(rotation=45) # 날짜 겹침 방지를 위해 회전
plt.grid(True, linestyle='--', alpha=0.7) # 격자 추가
plt.legend() # 범례 표시

# 4. 그래프 출력
plt.tight_layout()
plt.show()
```
