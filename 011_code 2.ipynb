```python
# 1. 분석에 필요한 라이브러리 불러오기
import pandas as pd

# 2. 실무형 데이터 불러오기 (정확한 파일명 확인!)
# 앞서 만든 '011 file_practice_data_2026.csv' 파일을 불러옵니다.
df = pd.read_csv('011 file_practice_data_2026.csv')

# 3. '특성(Feature)'과 '정답(Label)' 지정하기
# AI가 학습할 재료 (기온과 미세먼지를 보고 오존농도를 예측한다고 가정)
# 열 이름에 특수문자나 영문이 포함되어 있으므로 정확히 입력해야 합니다.
features = df[['기온(°C)', '미세먼지(PM10)']]

# AI가 맞혀야 할 목표 (정답)
# '오존농도' 대신 실제 열 이름인 '오존농도(Target)'를 사용합니다.
label = df['오존농도(Target)']

print("[STEP 1] AI의 학습 재료 (Feature) 확인:")
display(features.head())

print("\n[STEP 2] AI가 맞혀야 할 정답 (Label) 확인:")
display(label.head())

# 4. 데이터 상태 최종 점검 (기획안 내용 반영)
print("\n[STEP 3] 데이터 품질 점검 (결측치 및 타입 확인):")
df.info()
```
