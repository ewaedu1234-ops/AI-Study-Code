```python
def convert_currency(amount, currency_code):
    # 미리 정의된 환율 데이터 (실제로는 API에서 가져온 데이터)
    rates = {"KRW": 1350.5, "EUR": 0.92, "JPY": 155.2}

    if currency_code in rates:
        result = amount * rates[currency_code]
        return result
    else:
        return "지원하지 않는 통화입니다."
```

```python
# 단 한 줄로 환전 결과 확인!
print("500달러 환산 결과:")
print(f"원화(KRW): {convert_currency(500, 'KRW'):,.0f}원")
print(f"유로(EUR): {convert_currency(500, 'EUR')}유로")
print(f"엔화(JPY): {convert_currency(500, 'JPY')}엔")
```
