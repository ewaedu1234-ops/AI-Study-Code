```python
import requests

# 날씨 API 키
WEATHER_KEY = "4ae57fd056a83f59e7561791b26b4d3a"
url = "https://api.openweathermap.org/data/2.5/weather"

params = {
    "q":     "Seoul",
    "appid": WEATHER_KEY,
    "units": "metric",
    "lang":  "kr"
}

# 데이터 가져오기
response = requests.get(url, params=params)
print("상태 코드:", response.status_code)

# 값 꺼내기
data     = response.json()
기온     = data["main"]["temp"]
날씨상태  = data["weather"][0]["description"]
습도     = data["main"]["humidity"]
도시     = data["name"]

# 결과 출력
print("=" * 40)
print("     오늘의 날씨 요약")
print("=" * 40)
print(f"도시:  {도시}")
print(f"기온:  {기온}℃")
print(f"날씨:  {날씨상태}")
print(f"습도:  {습도}%")
print("=" * 40)
```
