```python
import csv

# 기록할 데이터 (리스트 형태)
new_data = ["2026-05-15", "서울", 23.5, 38, "맑음"]

with open("weather_data.csv", "a", encoding="utf-8-sig", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(new_data) # 리스트가 한 행으로 쏙 들어갑니다.
```

```python
import pandas as pd

# 데이터 불러오기
df = pd.read_csv("weather_data.csv")
print(df) # 표 형태로 예쁘게 출력됩니다.
```

```python
# 지난 7일간의 평균 기온은?
avg_temp = df["기온"].mean()
print(f"평균 기온: {avg_temp:.1f}도")

# 가장 습도가 높았던 날은 언제일까?
max_humidity = df.loc[df["습도"].idxmax()]
print(f"최고 습도 날짜: {max_humidity['날짜']} ({max_humidity['습도']}%)")
```
