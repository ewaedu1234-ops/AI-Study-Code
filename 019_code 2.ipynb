```python
import requests

# 환율 API 키
EXCHANGE_KEY = "3b1545d51227888fd69123c1"
url = f"https://v6.exchangerate-api.com/v6/{EXCHANGE_KEY}/latest/USD"

# 데이터 가져오기
response = requests.get(url)
print("상태 코드:", response.status_code)

# 환율 꺼내기
data = response.json()
환율_원화 = data["conversion_rates"]["KRW"]
환율_유로 = data["conversion_rates"]["EUR"]
환율_엔화 = data["conversion_rates"]["JPY"]
기준통화   = data["base_code"]

# 환산할 달러 금액 (원하는 숫자로 바꿀 수 있습니다)
내_달러   = 100
원화_환산 = 내_달러 * 환율_원화

# 결과 출력
print("=" * 45)
print("     오늘의 환율 계산기")
print("=" * 45)
print(f"1 {기준통화} = {환율_원화:,.1f} 원")
print(f"1 {기준통화} = {환율_유로} 유로")
print(f"1 {기준통화} = {환율_엔화} 엔")
print("-" * 45)
print(f"내 달러: {내_달러} USD")
print(f"원화 환산: {원화_환산:,.0f} 원")
print("=" * 45)
```
